```python
df = pd.read_csv('heart-disease.csv')
```

```python
df
```

```python
df.info()
```

```python
#create a histogram
df['age'].plot.hist()
```

```python
df.plot.hist(figsize=(15,8),subplots=True);
```

```python
#pyplot vs matplotlib oo method
```

```python
over_50 = df[df['age']>50]
over_50
```

```python
#pyplot method
over_50.plot(kind='scatter',
             x='age',
             y='chol',
             c='target')
```

```python
# oo methid
fig,ax = plt.subplots(figsize=(10,6))
over_50.plot(kind='scatter',
             x='age',
             y='chol',
             c='target',
             ax=ax);

ax.set_xlim([45,100]) 
```

```python
over_50.target.values
```

```python
fig,ax = plt.subplots(figsize=(10,6))

scatter = ax.scatter(x=over_50['age'],
                     y=over_50['chol'],
                     c=over_50['target'])

#customize the plot
ax.set(title='heart disease and cholestrol level',
       xlabel='age',
       ylabel='cholestrol')

ax.legend(*scatter.legend_elements(),title='Target')

#add a horizontal line
ax.axhline(over_50['chol'].mean(),
           linestyle='--')
```

```python
#subplot of chol, age, thalach
fig, (ax0,ax1) = plt.subplots(nrows=2,ncols=1,
                              figsize=(15,7),
                              sharex=True)
scatter = ax0.scatter(x=over_50['age'],
                      y=over_50['chol'],
                      c=over_50['target'])
ax0.set(title='heart disease and cholestrol levels',
        ylabel='chol')
ax0.legend(*scatter.legend_elements(),title='Target')
ax0.axhline(y=over_50['chol'].mean(),
            linestyle='--');

scatter = ax1.scatter(x=over_50['age'],
                      y=over_50['thalach'],
                      c=over_50['target'])
ax1.set(title='heart disease and maxh heart rate',
        xlabel='age',
        ylabel='thalach')
ax1.legend(*scatter.legend_elements(),title='Target')
ax1.axhline(y=over_50['thalach'].mean(),
            linestyle='--');

#add a title to the figure
fig.suptitle('Heart Disease Analysis',fontsize=16,fontweight='bold')
```
